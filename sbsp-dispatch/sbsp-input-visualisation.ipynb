```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

day_ahead_prices_df = pd.read_csv('../data/day_ahead_prices_all_domains_20150101-20241231.csv', index_col=0, parse_dates=True)

day_ahead_prices_df.replace(-10000, np.nan, inplace=True)
day_ahead_prices_df.head()
```

```python
plt.figure(figsize=(20,8))
day_ahead_prices_df.boxplot()
plt.xticks(rotation=90)
plt.title('Boxplot of Day-Ahead Prices by Market')
plt.ylabel('Price (â‚¬/MWh)')
plt.grid(True)
plt.show()
```
